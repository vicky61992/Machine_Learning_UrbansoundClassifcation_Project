```python
import sys
import os
import IPython
import math
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
%matplotlib inline

import random
from datetime import datetime
import helpers
from tensorflow.keras import backend as keras_backend
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, SpatialDropout2D, Activation, Conv2D, MaxPooling2D, BatchNormalization, GlobalAveragePooling2D, LeakyReLU, Dropout,Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import ModelCheckpoint 
from tensorflow.keras.regularizers import l2


from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, precision_recall_fscore_support, confusion_matrix
import warnings
warnings.filterwarnings('ignore')
```

```python
X_train = pd.read_csv(r'C:\Users\VSBAG\Desktop\DSE_Milan\3rd_sem_subject\Machine Learning\Project\Sound_classification\train_data2.csv', delimiter=',')
Y_train = pd.read_csv(r'C:\Users\VSBAG\Desktop\DSE_Milan\3rd_sem_subject\Machine Learning\Project\Sound_classification\train_labels2.csv', delimiter=',')
X_test = pd.read_csv(r'C:\Users\VSBAG\Desktop\DSE_Milan\3rd_sem_subject\Machine Learning\Project\Sound_classification\test_data2.csv', delimiter=',')
Y_test = pd.read_csv(r'C:\Users\VSBAG\Desktop\DSE_Milan\3rd_sem_subject\Machine Learning\Project\Sound_classification\test_labels2.csv', delimiter=',')
```

```python
from sklearn import preprocessing
x_train = preprocessing.normalize(X_train)
y_train = preprocessing.normalize(Y_train)
x_test = preprocessing.normalize(X_test)
y_test = preprocessing.normalize(Y_test)
x_train_norm = pd.DataFrame(x_train)
y_train_norm = pd.DataFrame(y_train)
x_test_norm = pd.DataFrame(x_test)
y_test_norm = pd.DataFrame(y_test)
x_train_norm.shape,y_train_norm.shape,x_test_norm.shape,y_test_norm.shape
```

```python
#converting to one hot
y_train = to_categorical(y_train_norm, num_classes=10)
y_test = to_categorical(y_test_norm, num_classes=10)
y_train.shape,y_test.shape
```

```python
#reshaping to 2D 
x_train=np.reshape(x_train,(x_train_norm.shape[0], 40,10))
x_test=np.reshape(x_test,(x_test_norm.shape[0], 40,10))
x_train.shape,x_test.shape
```

```python
#forming model
model=Sequential()

#adding layers and forming the model
model.add(Conv2D(64,kernel_size=4,strides=2,padding="Same",activation="relu",input_shape=(40,10,1)))
model.add(MaxPooling2D(padding="same"))

model.add(Conv2D(128,kernel_size=4,strides=2,padding="same",activation="relu"))
model.add(MaxPooling2D(padding="same"))

model.add(Conv2D(128,kernel_size=4,strides=2,padding="same",activation="relu"))
model.add(MaxPooling2D(padding="same"))
model.add(Dropout(0.3))

model.add(Flatten())

model.add(Dense(256,activation="relu"))
model.add(Dropout(0.3))

model.add(Dense(512,activation="relu"))
model.add(Dropout(0.3))

model.add(Dense(10,activation="softmax"))
```

```python
#compiling
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
#training the model
model.fit(x_train,y_train,batch_size=50,epochs=150,validation_data=(x_test,y_test))
```

```python
#train and test loss and scores respectively
train_loss_score=model.evaluate(x_train,y_train)
test_loss_score=model.evaluate(x_test,y_test)
print(train_loss_score)
print(test_loss_score)
```
